### Importing libraries

```python
import numpy as np 
%matplotlib inline 
import matplotlib.pyplot as plt
```

### Read the image

```python
image = cv2.imread("D:\Internships\LetsGrowMore\Task3-Image pencil sketch\German-shepherd.jpg")
```

```python
plt.figure(num = 1, figsize=(10,10))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image', fontsize=26)
plt.show()
```

### Convert to gray scale

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.figure(num=2,figsize=(10,10))
plt.imshow(gray_image,cmap='gray')
plt.title("gray image")
plt.show()
```

### Invert the grayscale image

```python
inverted_image = 255 - gray_image
plt.figure(num=3,figsize=(10,10))
plt.imshow(inverted_image,cmap='gray')
plt.title("inverted image")
plt.show()
```

### Blur image using Gaussian Blur Function

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
plt.figure(num=4,figsize=(10,10))
plt.imshow(blurred,cmap='gray')
plt.title("blurred image")
plt.show()
```

### Invert the blurred image

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
plt.figure(num=5,figsize=(10,10))
plt.imshow(pencil_sketch,cmap='gray')
plt.title("Pencil sketch")
plt.show()
```

### Original image

```python
plt.figure(num=6,figsize=(10,10))
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.title("Original image")
plt.show()
```

### Final Pencil sketch of the image

```python
plt.figure(num=7,figsize=(10,10))
plt.imshow(pencil_sketch,cmap='gray')
plt.title("Pencil sketch")
plt.show()
```
