# Data Visualization

```python
fig,axes = plt.subplots(1, 1, figsize = (15, 15))
sns.heatmap(df.corr(), annot =True)
plt.show()
```

# Attacking city wise

```python
plt.figure(figsize=(15, 10))
x = df['city'].value_counts().index[1:10]
y = df['city'].value_counts().values[1:10]
sns.barplot(x,y,palette = 'Greys_r')
plt.title('Top 20 Cities Affected By Terrorism',size=20)
plt.xlabel('Cities', size = 16)
plt.ylabel('Count', size = 16)
plt.xticks(size = 15)
plt.show()
```

# The most affected city is Baghdad

# Attack Country wise

```python
plt.figure(figsize = (15, 10))
x = df['Country'].value_counts().index[:10]
y = df['Country'].value_counts().values[:10]
sns.barplot(x,y,palette='deep')
plt.title(' Top 10 Countries Affected By Terrorism',size = 20)
plt.xlabel('Countries', size = 15)
plt.ylabel('Count', size = 15)
plt.xticks(size = 13)
plt.show()
```

# The most affected country is Iraq

# Attack year wise 

```python
plt.figure(figsize = (15, 10))
sns.countplot(data = df, x = 'Year',palette='Blues')
plt.xticks(rotation = 90)
plt.xlabel("Year",size = 20)
plt.ylabel("Count",size = 20)
plt.show()
```

# In The Year 2014 has the maximum terror attacks

# Attack Region wise

```python
plt.figure(figsize = (20, 10))
sns.countplot(data = df, x = 'Region') 
plt.xticks(size = 20)
plt.xlabel("REGION",size = 10)
plt.ylabel("COUNT",size = 10) 
plt.xticks(rotation = 90)
plt.show()
```

# Terrorist groups

```python
df['Group'].value_counts().to_frame().drop('Unknown').head(10).plot(kind = 'bar',color = 'Pink', figsize = (15, 10))
plt.title("Top 15 terrorist group attack")
plt.xlabel("terrorist group name")
plt.ylabel("Attack number")
plt.show()
```

# Talibans has maximum atacks
